```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
st = time.time()
x=print(func2(2, 4))
end = time.time()
# This function finds one number raise to power another using multiplication
print(f"elapsed time : {(end - st)/1000} ms")
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(lst):
  even_count = 0
  odd_count = 0
  for x in lst:
    if x % 2 == 0:
      even_count += 1
    else:
      odd_count += 1
  return even_count, odd_count
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(lst, target):
  for i in range(len(lst)):
    if lst[i] == target:
      return i
  return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
array = [1, 3, 5, 7, 9]
st = time.time()
print(linear_Search(array, 7))
end = time.time()
# This function finds the index of target element
print(f"elapsed time : {(end - st)/1000} ms")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(lst, target):
  i = 0
  j = len(lst) - 1
  while i <= j:
    mid = (i + j) // 2
    if lst[mid] == target:
      return mid
    elif lst[mid] < target:
      i = mid + 1
    else:
      j = mid - 1
  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9,15,16,14,45]
st = time.time()
print(Binary_Search(arr, 15))
end = time.time()
# This function finds the index of target element
print(f"elapsed time : {(end - st)/1000} ms")
```
