# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
import math as m
import numpy as np
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
def even_oddcounts(numbers):
  even_count = 0
  odd_count = 0
  for n in numbers:
    if n % 2 == 0:
      even_count += 1
    else:
      odd_count += 1
  return even_count, odd_count


val3 = list(range(1, 11))
time3 = []

for n in val3:
  data = list(range(n))
  even_oddcounts(data)
  time3.append(n)

plt.plot(val3, time3)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for even odd counts")
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
def binary_search(arr, k):
  i = 0;
  j = len(arr) - 1
  while(i < j):
      mid = i + (j -i)//2
      if arr[mid] == k:
        return True
      if arr[mid] < k:
        i = mid + 1
      else:
        j = mid - 1

val4 = list(range(1, 11))
time4 = []

for n in val4:
  data = list(range(n))
  binary_search(data, (n%10)*2)
  time4.append(m.log(n, 2))

plt.plot(val4, time4)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(logn) Logarithmic Time Complexity for binary search")
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in Polynomial time. plot the graph as well, similar to above example
```

```python
def print_2D(arr):
  for r in arr:
    for elem in r:
      pass


val6 = list(range(1, 11))
time6 = []

for n in val6:
  arr = np.arange(1, n*n+1).reshape(n, n)
  print_2D(arr)
  time6.append(n**2)

plt.plot(val6, time6)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n^2) Polynomial Time Complexity for printing 2D array")
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in Exponential time. plot the graph as well, similar to above example
def fibbonacchi(n):
  if n == 0 or n == 1:
    return n
  return fibbonacchi(n-1) + fibbonacchi(n-1)

val7 = list(range(1, 11))
time7 = []
for n in val7:
  fibbonacchi(n)
  time7.append(2**n)


plt.plot(val7, time7)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(2^n) Exponential Time Complexity to find fibbonacchi term")
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Polynomial Time Complexity O(n^2)
```

```python
def partition(arr, low, high):
  pivot_value = arr[low]

  count_smaller = 0
  for k in range(low + 1, high + 1):
    if arr[k] < pivot_value:
      count_smaller += 1

  pivind = low + count_smaller
  arr[low] , arr[pivind] = arr[pivind], arr[low]

  i = low
  j = high

  while i < pivind and j > pivind:
    while i < pivind and arr[i] <= pivot_value:
      i += 1

    while j > pivind and arr[j] >= pivot_value:
      j -= 1

    if i < pivind and j > pivind:
      arr[i] , arr[j] = arr[j], arr[i]
      i += 1
      j -= 1
  return pivind

def quick_sort(arr, i, j):
  if i < j:
    piv = partition(arr, i, j)
    quick_sort(arr, i, piv-1)
    quick_sort(arr, piv+1, j)


val8 = list(range(1, 11))
time8 = []
for n in val8:
  arr = np.random.randint(1, 100, (n))
  quick_sort(arr, 0, n-1)
  time8.append(n**2)


plt.plot(val8, time8, label='n^2')
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n^2) Polynomial Time Complexity of quick sort")
plt.legend()
plt.show()
```
